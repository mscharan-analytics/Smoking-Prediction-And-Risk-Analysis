## ML Preprocessing

### Pipeline Creation

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline, make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer, make_column_selector

from sklearn import set_config
set_config(display='diagram')

cat_columns = ['sex', 'hear_left', 'hear_right', 'urine_protein']
num_columns = [
    'age', 'height', 'weight', 'waistline', 'sight_left', 'sight_right',
    'SBP', 'DBP', 'BLDS', 'tot_chole', 'HDL_chole', 'LDL_chole',
    'triglyceride', 'hemoglobin', 'serum_creatinine', 'SGOT_AST',
    'SGOT_ALT', 'gamma_GTP'
]

num_pipeline = Pipeline([
        ('imputer', SimpleImputer(strategy='median')),
        ('scaler', StandardScaler())
    ])

cat_pipeline = Pipeline([
        ('imputer', SimpleImputer(strategy='most_frequent')),
        ('cat_encoder', OneHotEncoder(sparse_output=False))
    ])

prep_pipeline = ColumnTransformer([
    ('num', num_pipeline, num_columns),
    ('cat', cat_pipeline, cat_columns)
], verbose_feature_names_out=False)
prep_pipeline.set_output(transform='pandas')

prep_pipeline
```

Purpose:

The above code defines a data preprocessing pipeline using scikit-learn for a machine learning task. The pipeline is designed to handle both numerical and categorical features, applying appropriate transformations for each.

## Feature Selection

Feature selection is a critical step in machine learning aimed at improving model performance and interpretability. It addresses the fact that not all input features contribute equally to a model's predictive power, and some may introduce noise or redundancy. By carefully selecting a subset of relevant features, feature selection mitigates overfitting, enhances generalization to new data, and simplifies model interpretation. This streamlined set of features not only improves efficiency by reducing computational requirements but also aids in extracting meaningful insights. Overall, feature selection strategically optimizes machine learning models for better efficiency, interpretability, and generalization.

We will be using the following feature selection methods:
1. SelectKBest
2. SelectFromModel
3. Recursive Feature Elimination
4. Sequential Feature Selection : Forward and Backward

### SelectKBest

```python
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression

feature_selection_pipeline = Pipeline([
    ('prep', prep_pipeline),
    ('select', SelectKBest(score_func=f_classif)),
    ('model', LogisticRegression(random_state=42))
])

param_grid = {
    'select__k': range(1, X_train.shape[1] + 1)
}

grid_search = GridSearchCV(feature_selection_pipeline, param_grid, cv=5, scoring='balanced_accuracy')
grid_search.fit(X_train, y_train)

cv_res = pd.DataFrame(grid_search.cv_results_).sort_values(by='mean_test_score', ascending=False)
display(cv_res.filter(regex='(^param_|mean_test_score)', axis=1))

best_model = grid_search.best_estimator_
selected_features = best_model["prep"].get_feature_names_out()[best_model["select"].get_support()]
print(f'The selected features are {selected_features}')
```

###SelectFromModel

```python
from sklearn.feature_selection import SelectFromModel, SelectKBest, f_classif
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression

lasso_select_model = Lasso()
select_from_model_pipeline = Pipeline([
    ('transformer', prep_pipeline),
    ('select', SelectFromModel(estimator=lasso_select_model)),
    ('model', LogisticRegression(random_state=42))
])

param_grid = {
    'select__estimator__alpha': [0.01, 0.1, 1.0, 10, 100],
    'select__threshold': ['mean', 'median']
}

grid_search = GridSearchCV(select_from_model_pipeline, param_grid, cv=5, scoring='balanced_accuracy')
grid_search.fit(X_train, y_train)

cv_res = pd.DataFrame(grid_search.cv_results_).sort_values(by='mean_test_score', ascending=False)
display(cv_res.filter(regex='(^param_|mean_test_score)', axis=1))

best_model = grid_search.best_estimator_
selected_features = best_model["transformer"].get_feature_names_out()[best_model["select"].get_support()]
num_selected_features = len(selected_features)
print(f'The selected features are {selected_features}')
print(f'Number of selected features: {num_selected_features}')
```

###Recursive Feature Elimination RFECV

```python
from sklearn.feature_selection import RFECV
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline

rfecv = RFECV(LogisticRegression(random_state=42, max_iter=1000), scoring='balanced_accuracy')

rfecv_pipeline = Pipeline([
    ('transformer', prep_pipeline),
    ('select', rfecv),
    ('model', LogisticRegression(random_state=42))
])

rfecv_pipeline.fit(X_train, y_train)

print('Optimal number of features:', rfecv.n_features_)
print('The selected features are:\n', rfecv_pipeline['transformer'].get_feature_names_out()[rfecv.support_])

plt.figure(figsize=(10, 6))
plt.title("RFECV - Optimal Number of Features")
plt.xlabel("Number of Features Selected")
plt.ylabel("Performance of Balanced Accuracy")
mean_test_score = rfecv.cv_results_['mean_test_score']
plt.plot(range(1, len(mean_test_score ) + 1), mean_test_score)
```

###Sequential Feature Selection

**Forward Selection**

```python
from mlxtend.feature_selection import SequentialFeatureSelector
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline

sfs = SequentialFeatureSelector(LogisticRegression(random_state=42),
                                  k_features="best",
                                  scoring='balanced_accuracy',
                                  forward=True) # for Forward

sfs_pipeline = Pipeline([
    ('transformer', prep_pipeline),
    ('select', sfs),
    ('model', LogisticRegression(random_state=42))
])

sfs_pipeline.fit(X_train, y_train)

sfs_result = pd.DataFrame.from_dict(sfs.get_metric_dict()).T
display(sfs_result.sort_values(by='avg_score', ascending=False))
```

**Backward Selection**

```python
from mlxtend.feature_selection import SequentialFeatureSelector

sbs = SequentialFeatureSelector(LogisticRegression(random_state=42),
                                  k_features="best",
                                  scoring='balanced_accuracy',
                                  forward=False) # for Backward

sbs_pipeline = Pipeline([
    ('transformer', prep_pipeline),
    ('select', sbs),
    ('model', LogisticRegression(random_state=42))
])

sbs_pipeline.fit(X_train, y_train)

filter = sbs_pipeline.named_steps['transformer'].get_feature_names_out()

sbs_result = pd.DataFrame.from_dict(sbs.get_metric_dict()).T
display(sbs_result.sort_values(by='avg_score', ascending=False))
```

#### Choosing the best Feature selection model

Method selected : Backward SFS

Reason:

After conducting feature selection using various methods, including SelectKBest and SelectFromModel, it was observed that the first two methods yielded suboptimal results, providing meaningless features and the fewest number of selected features. On the other hand, the Sequential Feature Selection techniques, specifically Forward selection, demonstrated the best performance, achieving the highest scores with a selection of 11 features. However, despite its success, the Backward selection method is preferred for our final feature set. The rationale behind choosing Backward selection lies in its ability to iteratively eliminate the least significant features, starting from the complete set. This method aligns with a conservative approach, ensuring that the retained features collectively contribute the most to the model's predictive power. As a final decision, we have opted to proceed with selecting 15 key features, leveraging the insights gained from the feature selection process.

## Defining Metric Calculation function

```python
from sklearn.metrics import accuracy_score, balanced_accuracy_score, confusion_matrix, classification_report, roc_curve, auc
import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay, RocCurveDisplay
from sklearn.metrics import make_scorer
from sklearn.model_selection import cross_val_score

def evaluate_classification_model(model, X_train, y_train, X_test, y_test, cost_weight_fp=1, cost_weight_fn=10, cv_folds=5):
    # Fit the model
    model.fit(X_train, y_train)

    # Predicting label
    y_pred = model.predict(X_test)

    #Accuracy metrics
    accuracy = accuracy_score(y_test, y_pred)
    balanced_accuracy = balanced_accuracy_score(y_test, y_pred)

    #Confusion Matrix
    cm = confusion_matrix(y_test, y_pred)
    ConfusionMatrixDisplay(cm, display_labels=model.classes_).plot(values_format='d')
    plt.title('Confusion Matrix')
    plt.show()

    #Classification Report
    report = classification_report(y_test, y_pred, digits=3)
    print('Classification Report:')
    print(report)

    #Calculate ROC Curve and AUC
    y_probs = model.predict_proba(X_test)[:, 1]
    fpr, tpr, thresholds = roc_curve(y_test, y_probs)
    roc_auc = auc(fpr, tpr)

    # Plot ROC Curve
    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic (ROC) Curve')
    plt.legend(loc='lower right')
    plt.show()

    #Cost calculation
    def custom_cost(y_true, y_pred):
        cm = confusion_matrix(y_true, y_pred)
        return cm[1, 0] * cost_weight_fn + cm[0, 1] * cost_weight_fp

    cost_scorer = make_scorer(custom_cost)

    #Cross-validation for cost
    costs = cross_val_score(model, X_train, y_train, cv=cv_folds, scoring=cost_scorer)
    avg_cost = costs.mean()

    # Print results
    print(f'Balanced accuracy: {balanced_accuracy:.3f}')
    print(f'The average cost is {avg_cost:.1f}.')
```

```python
clf_full = DecisionTreeClassifier()

prune = clf_full.cost_complexity_pruning_path(X_train, y_train)
ccp_alphas, impurities = prune.ccp_alphas, prune.impurities # default impurity is Gini
plt.plot(ccp_alphas, impurities, marker='o', drawstyle='steps-post')
plt.xlabel('Alphas'); plt.ylabel('Total impurity');

print(f'There are {ccp_alphas.shape[0]} alpha values.')
```

When $Î±=0$, pruning is absent, and the leaves predominantly contain one of the two classes, resulting in a low Gini impurity index. Conversely, at the maximum $Î±$ value, the Gini impurity reflects that of the original training data without any splits, indicating a high Gini impurity index. Additionally, visualizing the number of nodes and the depth of the tree against varying $Î±$ allows us to observe how the tree's size changes with different $Î±$ values.

```python
clfs = []
for ccp_alpha in ccp_alphas:
    clf_i = DecisionTreeClassifier(random_state=0, ccp_alpha=ccp_alpha)
    clf_i.fit(X_train, y_train)
    clfs.append(clf_i)

node_counts = [clf_i.tree_.node_count for clf_i in clfs]
depth = [clf_i.tree_.max_depth for clf_i in clfs]

fig, ax = plt.subplots(2, 1, dpi=100)
ax[0].plot(ccp_alphas, node_counts, marker="o", drawstyle="steps-post")
ax[0].set_xlabel("alpha")
ax[0].set_ylabel("number of nodes")
ax[0].set_title("Number of nodes vs alpha")
ax[1].plot(ccp_alphas, depth, marker="o", drawstyle="steps-post")
ax[1].set_xlabel("alpha")
ax[1].set_ylabel("depth of tree")
ax[1].set_title("Depth vs alpha")
fig.tight_layout()
```

The trees get smaller as $Î±$ increases. But which one of those 74 trees predicts best? We can apply it to the test data and check:

```python
train_scores = [clf_i.score(X_train, y_train) for clf_i in clfs]
test_scores = [clf_i.score(X_test, y_test) for clf_i in clfs]

fig, ax = plt.subplots(dpi=100)
ax.set_xlabel("alpha")
ax.set_ylabel("accuracy")
ax.set_title("Accuracy vs alpha for training and testing sets")
ax.plot(ccp_alphas, train_scores, marker="o", label="train", drawstyle="steps-post")
ax.plot(ccp_alphas, test_scores, marker="o", label="test", drawstyle="steps-post")
ax.legend()
plt.show()
```


The training accuracy consistently decreases as the parameter $Î±$ reduces, indicating a trend of diminishing performance as the tree becomes smaller. However, the test error exhibits its maximum value at an intermediate $Î±$.

It is essential to refrain from performing model selection based on test data. The test dataset should be reserved for the purpose of estimating generalization error. Instead, a more appropriate approach involves utilizing cross-validation in conjunction with grid search to identify the optimal $Î±$ among the specified values.
